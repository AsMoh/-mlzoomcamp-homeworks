### Split the data.

Shuffle the initial dataset, use seed 42.

Split your data in train/val/test sets, with 60%/20%/20% distribution.
Make sure that the target value ('median_house_value') is not in your dataframe.
Apply the log transformation to the median_house_value variable using the np.log1p() function.

```python
def split_df(df, test_size=0.2, se=42):
    n = len(df)
    n_val = int(n * 0.2)
    n_test = int(n * 0.2)
    n_train = n - n_val - n_test
    idx=np.arange(0,n)
    np.random.seed(se)
    np.random.shuffle(idx)
    df_train=df.iloc[idx[:n_train]]
    df_val=df.iloc[idx[n_train:n_train+n_val]]
    df_test=df.iloc[idx[n_train+n_val:]]
    
    df_train.reset_index( drop=True, inplace=True)
    df_val.reset_index( drop=True, inplace=True)
    df_test.reset_index( drop=True, inplace=True)
    
   
    y_train=df_train['median_house_value'].values
    y_val=df_val['median_house_value'].values
    y_test=df_test['median_house_value'].values
    
    del df_train['median_house_value']
    del df_val['median_house_value']
    del df_test['median_house_value']
    
    X_train=df_train
    X_val=df_val
    X_test=df_test
        
    return X_train,X_val, X_test, y_train,y_val,y_test
    
    
```

### Apply the log transformation to the median_house_value variable using the np.log1p() function.

```python
X_train,X_val, X_test, y_train,y_val,y_test= split_df(df,42)

y_train=np.log1p(y_train)
y_val=np.log1p(y_val)
y_test=np.log1p(y_test)

print(X_train.shape,X_val.shape, X_test.shape, y_train.shape,y_val.shape,y_test.shape)
```

```python
def dot(xi,w):
    res=0.0
    for i in range(0, len(xi)):
        res=res+ xi[i]*w[i]
    return res
               
```

```python
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    
    return w_full[0], w_full[1:]
```

```python
def RMSE(y_test, y_pridected):
    d=(y_test-y_pridected)**2
    mse=d.mean()
    return np.sqrt(mse)
```

### Question 4
Now let's train a regularized linear regression. </br>
For this question, fill the NAs with 0. </br>
Try different values of r from this list: [0, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10].  </br>
Use RMSE to evaluate the model on the validation dataset.  </br>
Round the RMSE scores to 2 decimal digits.  </br>
Which r gives the best RMSE? </br>
If there are multiple options, select the smallest r.  </br>

```python
def train_linear_regression_reg(X, y, r=0.001):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX = XTX + r * np.eye(XTX.shape[0])

    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    
    return w_full[0], w_full[1:]
```

```python
df_fillwith_0=prepare_df(df_train,0)

X_train,X_val, X_test, y_train,y_val,y_test= split_df(df_fillwith_0,42)
y_train=np.log1p(y_train)
y_val=np.log1p(y_val)
y_test=np.log1p(y_test)
```

```python
rms=[]
Ws=[]
for r in [0, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10]:
    b,W=train_linear_regression_reg(X_train,y_train, r)
    Ws.append((b,W))
    y_pridected=predict(X_val)
    rms.append(RMSE(y_val,y_pridected).round(6))
    
```

```python
print("r: ", [0, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 5, 10])
#print(Ws)
print("RMSE:" , rms)
```

```python
print(np.round(rms,2))
```
